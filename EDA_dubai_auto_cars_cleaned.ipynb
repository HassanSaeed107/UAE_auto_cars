## Outliers Visualization

```python
# Set style for seaborn plots
sns.set(style="whitegrid")

# Numerical variables with potential outliers
numerical_vars = ['price', 'kilometers', 'year', 'seating_capacity']

# Create box plots to visualize distributions and identify outliers
plt.figure(figsize=(12, 8))
for i, var in enumerate(numerical_vars, start=1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=df[var], color='skyblue')
    plt.title(f'Boxplot of {var}')
    plt.xlabel('')
plt.tight_layout()
plt.show()
```

### Detecting Outlier

```python
# Assessing outliers in numerical variables
numerical_vars = ['price', 'kilometers', 'year', 'seating_capacity']

for var in numerical_vars:
    # Calculate the 25th and 75th percentiles
    q25, q75 = df[var].quantile(0.25), df[var].quantile(0.75)

    # Calculate the interquartile range (IQR)
    iqr = q75 - q25

    # Define the upper and lower bounds for outliers detection
    upper_bound = q75 + 1.5 * iqr
    lower_bound = q25 - 1.5 * iqr

    # Identify potential outliers
    potential_outliers = df[(df[var] < lower_bound) | (df[var] > upper_bound)]

    # Assess outliers based on the provided questions
    print(f"Assessment of outliers in '{var}':")
    print(f"Number of potential outliers: {len(potential_outliers)}")

```

### Removing Outlier

```python
# Remove outliers from the dataset
cleaned_df = df.copy()  # Make a copy of the original dataframe
for var in numerical_vars:
    q25, q75 = df[var].quantile(0.25), df[var].quantile(0.75)
    iqr = q75 - q25
    upper_bound = q75 + 1.5 * iqr
    lower_bound = q25 - 1.5 * iqr
    cleaned_df = cleaned_df[(cleaned_df[var] >= lower_bound) & (cleaned_df[var] <= upper_bound)]
```

### Visualizing After removing Outliers

```python
# Set up the figure and axes
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# Plot histograms for 'price', 'kilometers', and 'year' after removing outliers
sns.histplot(df['price'], ax=axes[0], kde=True, color='skyblue')
axes[0].set_title('Distribution of Price (After Removing Outliers)')
axes[0].set_xlabel('Price')

sns.histplot(df['kilometers'], ax=axes[1], kde=True, color='salmon')
axes[1].set_title('Distribution of Kilometers (After Removing Outliers)')
axes[1].set_xlabel('Kilometers')

sns.histplot(df['year'], ax=axes[2], kde=True, color='green')
axes[2].set_title('Distribution of Year (After Removing Outliers)')
axes[2].set_xlabel('Year')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
